```python
Exportar librerias 

import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import joblib
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
Abrir archivo CSV

file = os.path.join('C:/Users/joaqu/Downloads', 'DataAceites.csv')
df1 = pd.read_csv(file)
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
Renombrando columnas(ordenando)

df1 = df1.rename(columns={
	'Fierro': 'ID',
	'Cromo': 'Fierro',
	'Aluminio': 'Cromo',
	'Cobre': 'Aluminio',
	'Plomo': 'Cobre',
	'Nickel': 'Plomo',
	'Plata': 'Nickel',
	'Estano': 'Plata',
	'Titanio': 'Estaño',
	'Vanadio': 'Titanio',
	'Cadmio': 'Vanadio',
	'Manganeso': 'Cadmio',
	'Sodio': 'Manganeso',
	'Potasio': 'Sodio',
	'Silicio': 'Potasio',
	'Zinc': 'Silicio',
	'Bario': 'Zinc',
	'Boro': 'Bario',
	'Calcio': 'Boro',
	'Molibdeno': 'Calcio',
	'Magnesio': 'Molibdeno',
	'Fosforo': 'Magnesio',
	'v 40': 'Fósforo',
	'Contenido agua': 'v 40',
	'Índice PQ': 'Contenido agua',
	'Oxidacion': 'Índice PQ',
	'Partículas > 4um': 'Oxidación',
	'Partículas > 6um': 'Partículas > 4um',
	'Partículas > 14um': 'Partículas > 6um',
	'Family': 'Partículas > 16um',
	'Estado': 'Familia',
	'Diagnostic': 'Estado',
	'Comentario': 'Diagnóstico',
	'Recomendacion': 'Comentario',
	'Unnamed: 34': 'Recomendación',
})
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
Eliminamos las columnas que no son necesarias

df1 = df1.drop([
	'Estado',
	'Recomendación',
	'Familia',
	'Comentario',
	'ID',
], axis='columns')
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
Reemplazamos los datos de letras a numeros para clasificarlos eficientemente 

df1 = df1
cond = ((df1['Estado de salud'].eq('Component Wear')))
df1['Estado de salud'].mask(cond, other=1, inplace=True)

cond = ((df1['Estado de salud'].eq('Silica ISO 4406')))
df1['Estado de salud'].mask(cond, other=2, inplace=True)

cond = ((df1['Estado de salud'].eq('Oil Contamination')))
df1['Estado de salud'].mask(cond, other=3, inplace=True)

cond = ((df1['Estado de salud'].eq('Silica and Wear ')))
df1['Estado de salud'].mask(cond, other=4, inplace=True)

cond = ((df1['Estado de salud'].eq('Water Contamination ')))
df1['Estado de salud'].mask(cond, other=5, inplace=True)

cond = ((df1['Estado de salud'].eq('Normal')))
df1['Estado de salud'].mask(cond, other=0, inplace=True)

attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
Finalmente guardamos el archivo como CSV 

file = os.path.join('C:/Users/joaqu/Downloads', 'DataAceitesDepurado')
df1.to_csv(file)
attrs = {
	'var': 'empty',
	'type': 'empty',
}
[attrs]
'csv file saved'
```
