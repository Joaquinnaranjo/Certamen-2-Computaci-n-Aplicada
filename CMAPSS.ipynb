```python
# Código general - importar librerías y función para preparar datos

from google.colab import drive
drive.mount('/content/drive')  # Solo si usas Colab; si no, borra estas dos líneas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def prepare_train_data(data, factor=0):
    df = data.copy()
    max_cycles = df.groupby('unit_number')['time_in_cycles'].max().reset_index()
    max_cycles.rename(columns={'time_in_cycles': 'max'}, inplace=True)
    df = df.merge(max_cycles, on='unit_number', how='left')
    df['RUL'] = df['max'] - df['time_in_cycles']
    df.drop(columns=['max'], inplace=True)
    return df[df['time_in_cycles'] > factor]
    
```

```python
# FD001

file_path_fd001 = "RUTA/A/TU/ARCHIVO/train_FD001.txt"  # Cambia aquí la ruta

fd_001 = pd.read_csv(file_path_fd001, sep=" ", header=None)
fd_001.drop(columns=[26, 27], inplace=True)

columns_fd001 = [
    "unit_number", "time_in_cycles", "setting_1", "setting_2", "TRA", "T2", "T24", "T30", "T50",
    "P2", "P15", "P30", "Nf", "Nc", "epr", "Ps30", "phi", "NRf", "NRc", "BPR",
    "farB", "htBleed", "Nf_dmd", "PCNfR_dmd", "W31", "W32"
]
fd_001.columns = columns_fd001

df1 = prepare_train_data(fd_001)

plt.figure(figsize=(20, 6))
df1.unit_number.value_counts().sort_index().plot.bar()
plt.xlabel("Id Simulado")
plt.ylabel("Cantidad de datos")
plt.title("Cantidad de registros por unidad simulada - FD001")
plt.show()
```
